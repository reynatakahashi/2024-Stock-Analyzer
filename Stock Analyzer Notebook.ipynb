#Getting Data

```python
import yfinance as yf
apple_stock = yf.Ticker("AAPL")
apple_data = apple_stock.history(period="1mo", start="2024-01-01", end="2024-12-31")

microsoft_stock = yf.Ticker("MSFT")
microsoft_data = microsoft_stock.history(period="1mo", start="2024-01-01", end="2024-12-31")

amazon_stock = yf.Ticker("AMZN")
amazon_data = amazon_stock.history(period="1mo", start="2024-01-01", end="2024-12-31")

visa_stock = yf.Ticker("V")
visa_data = visa_stock.history(period="1mo", start="2024-01-01", end="2024-12-31")

tesla_stock = yf.Ticker("TSLA")
tesla_data = tesla_stock.history(period="1mo", start="2024-01-01", end="2024-12-31")
```

#Viewing Apple Data


```python
import matplotlib.pyplot as plt

plt.plot(apple_data.index, apple_data["Close"])
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.title("Apple Stock Prices")
plt.show()

total_price_change=apple_data["Close"][-1]-apple_data["Close"][0]
print("The total price change is ", total_price_change)

total_percentage_change= (apple_data["Close"][-1]-apple_data['Close'][0])/apple_data["Close"][0]
print("The total percentage change is ", total_percentage_change)

lowest_price = min(apple_data["Close"])
print("The lowest price is ", lowest_price)

highest_price = max(apple_data["Close"])
print("The highest price is ", highest_price)

difference_of_prices= (max(apple_data["Close"]))-(min(apple_data["Close"]))
print("The difference in price is ", difference_of_prices)

last_price= apple_data["Close"][-1]
print("The last price is ", last_price)

percantage_difference= difference_of_prices/last_price
print("The difference in price in percentage is ", percantage_difference)

plt.plot(apple_data.index, apple_data["Volume"])
plt.xlabel("Date")
plt.ylabel("Volume")
plt.title("Apple Stock Volume")
plt.show()
```
